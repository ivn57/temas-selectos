```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Crear figura
plt.figure(figsize=(16, 8), dpi=150)

# Línea principal del precio de cierre
plt.plot(df.index, df["Close"], lw=2.5, color="#1f77b4", label="Precio de cierre")

# Sombreado del área bajo la curva para resaltar la tendencia
plt.fill_between(df.index, df["Close"], color="#1f77b4", alpha=0.1)

# Etiquetas y título
plt.xlabel("Fecha", fontsize=16, labelpad=10)
plt.ylabel("Precio ($)", fontsize=16, labelpad=10)
plt.title("Evolución del precio de cierre - Cemex", fontsize=22, fontweight="bold", pad=20)

# Formato del eje X para fechas
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # una etiqueta cada 3 meses
plt.xticks(rotation=45, ha="right")  # rotar las fechas para mejor lectura

# Cuadrícula más suave y elegante
plt.grid(which='major', color='gray', linestyle='--', alpha=0.3)
plt.grid(which='minor', color='gray', linestyle=':', alpha=0.1)

# Ajustes de ejes
plt.tick_params(labelsize=12, colors="dimgray")
plt.legend(fontsize=14, loc="upper left", frameon=False)

# Fondo de la figura más suave
plt.gca().set_facecolor("#f9f9f9")
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
plt.figure(figsize=(16,8),dpi=150)
plt.plot(df.index,df["Close"],lw=2.5,color="blue",label="Precio de cierre" )
min_idx=df["Close"].idxmin()
max_idx=df["Close"].idxmax()
min_val=df.loc[min_idx,"Close"]
max_val=df.loc[max_idx,"Close"]
plt.scatter(min_idx,min_val, color="red",s=80, label=f"Mínimo:${min_val:.2f}" )
plt.scatter(max_idx, max_val, color="green", s=80,label=f"Máximo:${max_val:.2f}")
plt.xlabel("Fecha",fontsize=18, labelpad=10)
plt.ylabel("Precio($)",fontsize=18,labelpad=10)
plt.title("Precio de cierre-WALMART",fontsize=22,fontweight="bold",pad=15)
plt.annotate(f"Mínimo\n${min_val:.2f}",
             xy=(min_idx, min_val),
             xytext=(-100,-40),textcoords="offset points",
             arrowprops=dict(facecolor="red",arrowstyle='->'),
             fontsize=14,color="red")
plt.annotate(f"Máximo\n${max_val:.2f}",
             xy=(max_idx,max_val),
             xytext=(40,40),
             textcoords="offset points",
             arrowprops=dict(facecolor="red",arrowstyle='->'),
             fontsize=14, color="green")
plt.grid(alpha=.03)
plt.tick_params(labelsize=14)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
print(f"precio de cierre más bajo:${min_val:.2f},fecha:{min_idx}")
print(f"precio de cierre más alto:${max_val:.2f},fecha:{max_idx}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.dates as mdates

# --- Supongamos que df ya tiene un índice de fechas y una columna "Close" ---

plt.style.use("seaborn-v0_8-darkgrid")  # estilo visual más moderno

# --- Cálculos estadísticos ---
mean_val = df["Close"].mean()
median_val = df["Close"].median()
std_val = df["Close"].std()

min_idx = df["Close"].idxmin()
max_idx = df["Close"].idxmax()
min_val = df.loc[min_idx, "Close"]
max_val = df.loc[max_idx, "Close"]

# --- Gráfica principal ---
fig, ax = plt.subplots(figsize=(16, 8), dpi=150)
ax.plot(df.index, df["Close"], lw=2.2, color="#0077b6", label="Precio de cierre")

# --- Puntos destacados ---
ax.scatter(min_idx, min_val, color="red", s=80, label=f"Mínimo: ${min_val:.2f}")
ax.scatter(max_idx, max_val, color="green", s=80, label=f"Máximo: ${max_val:.2f}")

# --- Líneas estadísticas ---
ax.axhline(mean_val, color="#ffb703", linestyle="--", lw=1.8, label=f"Media: ${mean_val:.2f}")
ax.axhline(median_val, color="#fb8500", linestyle=":", lw=1.8, label=f"Mediana: ${median_val:.2f}")
ax.axhline(mean_val + std_val, color="#219ebc", linestyle="--", lw=1, alpha=0.6, label=f"+1σ: ${mean_val + std_val:.2f}")
ax.axhline(mean_val - std_val, color="#219ebc", linestyle="--", lw=1, alpha=0.6, label=f"-1σ: ${mean_val - std_val:.2f}")

# --- Anotaciones ---
ax.annotate(f"Mínimo\n${min_val:.2f}",
            xy=(min_idx, min_val),
            xytext=(-80, -50),
            textcoords="offset points",
            arrowprops=dict(facecolor="red", arrowstyle='->'),
            fontsize=13, color="red")

ax.annotate(f"Máximo\n${max_val:.2f}",
            xy=(max_idx, max_val),
            xytext=(40, 40),
            textcoords="offset points",
            arrowprops=dict(facecolor="green", arrowstyle='->'),
            fontsize=13, color="green")

# --- Ejes y formato de fechas ---
ax.set_xlabel("Fecha", fontsize=16, labelpad=10)
ax.set_ylabel("Precio ($)", fontsize=16, labelpad=10)
ax.set_title("Evolución del Precio de Cierre - WALMART", fontsize=22, fontweight="bold", pad=15)

# Formateo de fechas (si el índice es tipo datetime)
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
plt.setp(ax.get_xticklabels(), rotation=45, ha="right")

# --- Ajustes visuales ---
ax.tick_params(labelsize=12)
ax.legend(fontsize=12, loc="best", frameon=True)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# --- Impresión de resumen estadístico ---
print("📊 Estadísticas del precio de cierre:")
print(f"  Precio mínimo: ${min_val:.2f} ({min_idx})")
print(f"  Precio máximo: ${max_val:.2f} ({max_idx})")
print(f"  Media: ${mean_val:.2f}")
print(f"  Mediana: ${median_val:.2f}")
print(f"  Desviación estándar: ${std_val:.2f}")
```
